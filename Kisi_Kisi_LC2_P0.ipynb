###**SQL**

#####1. Berapa banyak label unique dari bike_type?

````python
df_unik1 = client.query('''
SELECT COUNT(DISTINCT(bike_type))
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
''').to_dataframe()

df_unik1
````

#####2. Apa saja nama label unique dari tipe sepeda?

````python
# working area
df_unik = client.query('''
SELECT DISTINCT (bike_type) AS speda
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
''').to_dataframe()

df_unik
````

#####3. Berapa banyak total peminjaman sepeda pada tahun 2016 dan 2017?

````python
df_tahun = client.query('''
SELECT count(trip_id) as total
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
where EXTRACT(YEAR FROM start_time) = 2016 or EXTRACT(YEAR FROM start_time) = 2017
''').to_dataframe()


df_tahun
````

#####4. Dimana station yang paling banyak dikunjungi sebagai titik keberangkatan pada tahun 2016?

````python
# working area
df_stasiun = client.query('''
SELECT start_station_name as Nama_Stasiun, count (start_station_name) as Total_Kunjungan
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
where EXTRACT(YEAR FROM start_time) = 2016
group by start_station_name
order by Total_Kunjungan desc
''').to_dataframe()

df_stasiun
````

#####5. Berapa nilai minimum, maximum, dan rata-rata dari durasi perjalanan pada keseluruhan data?

````python
# working area
df_total = client.query('''
SELECT
AVG(duration_minutes) as rata_rata,
MAX(duration_minutes) as maksimal,
MIN(duration_minutes) as minimal
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
''').to_dataframe()

df_total
````

#####6. Berapa rata-rata durasi perjalanan dari subscriber_type `Student Membership` dan `Local31`?

````python
# working area
df_rata2 = client.query('''
SELECT subscriber_type,
AVG(duration_minutes) as rata_rata,
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE subscriber_type = "Student Membership" or subscriber_type = "Local31"
GROUP BY subscriber_type
''').to_dataframe()

df_rata2
````

#####7. Tampilkan entry data yang memiliki `duration_time` diatas 4.25 jam. (*Gunakan subquery untuk menjawab pertanyaan ini dengan meng-convert duration_time dari menit ke hour*).

````python
# working area


df_jam = client.query('''
SELECT *
FROM bigquery-public-data.austin_bikeshare.bikeshare_trips
WHERE (select duration_minutes/60 >= 4.25 as Durasi_Jam)

order by duration_minutes
limit 30
''').to_dataframe()



df_jam


# SELECT duration_minutes / 60 as durasi_jam
# FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
# where duration_minutes / 60 >= 4.25
#
````
